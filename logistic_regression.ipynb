```python
x = df.drop('diagnosis', axis=1)
y = df['diagnosis'].map({
    'B' : 0,
    'M' : 1
})
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    x, y,
    test_size=0.2,
    random_state=42,
    stratify=y
)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(x_train_sc, y_train)
```

```python
y_pred = model.predict(x_test_sc)
y_proba = model.predict_proba(x_test_sc)[:, 1]
```

```python
from sklearn.metrics import accuracy_score, roc_curve, confusion_matrix, recall_score, precision_score, roc_auc_score

cfm = confusion_matrix(y_test, y_pred)
acc = accuracy_score(y_pred, y_test)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
roc_score = roc_auc_score(y_test, y_pred)

print('Confusion Matrix :\n', cfm)
print('Accuracy Score :', acc)
print('Precision Score :', precision)
print('Recall Score  :', recall)
print('Roc Curve Score  :', roc_curve)
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_proba)

plt.plot(fpr, tpr, label = 'ROC Curve')
plt.plot([0, 1], [0, 1], linestyle = '--')
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('ROC CURVE')
plt.legend()
plt.show()
```

```python
custom_threshold = 0.3 # Be more sensitive. If you’re even 30% confident it’s malignant, flag it
y_pred_custom = (y_proba >= custom_threshold).astype(int)

from sklearn.metrics import precision_score, recall_score
print("Precision (0.3):", precision_score(y_test, y_pred_custom))
print("Recall (0.3):", recall_score(y_test, y_pred_custom))
```
